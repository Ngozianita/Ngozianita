```python
df = sns.load_dataset('titanic')
df.head()
```

```python
#2.COUNT PLOT: A count plot is helpful when dealing with categorical values. 
#It is used to plot the frequency of the different categories. 
#column sex contains categorical data in the titanic data, i.e., male and female.
sns.countplot(x='sex',data=df)
```

```python
#3. KDE Plot
#A Kernel Density Estimate (KDE) Plot is used to plot the distribution of continuous data.

sns.kdeplot(x = 'age' , data = df , color = 'black')
```

```python
sns.scatterplot(x='sepal_length', y ='petal_length' ,
data = df , hue = 'species')
```

```python
sns.displot(x = 'age',kde=True,bins = 5 , data =df)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1. subplot function can be used to draw multiple plots in one figure. The subplot() function takes three arguments that describes the layout of the figure.

#The layout is organized in rows and columns, which are represented by the first and second argument.

#The third argument represents the index of the current plot.

#plot 1:
x = np.array([10, 20, 30, 40])
y = np.array([5, 8, 10, 30])

plt.subplot(5, 2, 4)
plt.plot(x,y)

#plot 2:
x = np.array([1, 1, 2, 3])
y = np.array([10, 25, 30, 40])

plt.subplot(1, 9, 2)
plt.plot(x,y)

plt.show()
```

```python
#2. Grid function(): It is used to add grid lines to a plot


x = np.array([10, 20, 30, 40])
y = np.array([5, 8, 10, 30])

plt.title("WTF FELLOWSHIP '23")
plt.xlabel("Data Science Group C")
plt.ylabel("Sub Group 9")

plt.plot(x, y)

plt.grid()

plt.show()
```

```python
#3. The scatter() function: plots one dot for each observation. 
#It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis:

x = np.array([10, 20, 30, 40])
y = np.array([5, 8, 10, 30])
plt.scatter(x, y)
plt.show()
```

```python
#4.Label function():you can use the xlabel() and ylabel() functions to set a label for the x- and y-axis.

x = np.array([10, 20, 30, 40])
y = np.array([5, 8, 10, 30])
plt.plot(x, y)

plt.xlabel("Data science group C")
plt.ylabel("Sub Group 9")

plt.show()
```

```python
#5. Title function: you can use the title() function to set a title for the plot.


x = np.array([10, 20, 30, 40])
y = np.array([5, 8, 10, 30])
plt.plot(x, y)

plt.title("WTF FELLOWSHIP '23")
plt.xlabel("Data science group C")
plt.ylabel("Sub Group 9")

plt.show()

```

```python
#6. Marker Function: You can use the keyword argument marker to emphasize each point with a specified marker:

ypoints = np.array([5, 8, 10, 30])

plt.plot(ypoints, marker = '*')
plt.show()
```

```python
#7. Plot Function(): The plot() function is used to draw points (markers) in a diagram.

#By default, the plot() function draws a line from point to point.

#The function takes parameters for specifying points in the diagram.

#Parameter 1 is an array containing the points on the x-axis.

#Parameter 2 is an array containing the points on the y-axis.

#If we need to plot a line from (5, 20) to (10, 30), we have to pass two arrays [1, 8] and [3, 10] to the plot function.

xpoints = np.array([5, 20])
ypoints = np.array([10, 30])

plt.plot(xpoints, ypoints)
plt.show()
```
